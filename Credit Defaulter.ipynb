```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('german.data.txt', delim_whitespace=True, names=np.arange(1,22))
```

```python
df.head(20)
```

```python
sns.pairplot(df,hue=21)
```

```python
df.corr()
```

```python
df.info()
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
df=pd.get_dummies(df)
df.head()
#df.corr().describe().head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df.drop(21,axis=1))
scaled_data=scaler.transform((df.drop(21,axis=1)))
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=6)
pca.fit(scaled_data)
x_pca=pca.transform(scaled_data)
scaled_data.shape
x_pca.shape
plt.figure()
plt.scatter(x_pca[:,0],x_pca[:,1],c=df[21])
df[21].value_counts
```

```python
from sklearn.model_selection import train_test_split
```

```python
new_dataframe=pd.DataFrame(x_pca, columns=['c1','c2','c3','c4','c5','c6'])
new_dataframe['Target']=df[21]
new_dataframe.head()
#X=new_dataframe.drop(new_dataframe['Target'], axis=1)
#X=new_dataframe.drop(new_dataframe['Target'],axis=1)
X=new_dataframe.iloc[:, :-1].values
y=new_dataframe['Target']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
prediction=lg.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
```

```python
print(confusion_matrix(y_test,prediction))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=22)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
pred = dt.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
pred_rfc=rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred_rfc))
```
